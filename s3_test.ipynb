```python
import boto3, botocore, re, zlib, json
from pprint import pprint
import pandas as pd
from io import BytesIO
from gzip import GzipFile
```

```python
s3 = boto3.client(
    's3',
    aws_access_key_id="-------",
    aws_secret_access_key="-------")
```

```python
response = s3.list_buckets()
```

```python
paginator = s3.get_paginator('list_objects')
page_iterator = paginator.paginate(Bucket='cloudtrail-logs-cloudthat')
```

```python
targetFiles = []
targetRecords = []
nonTargetRecords = []
pattern = re.compile('.*_CloudTrail_.*')
runCount = 0

for page in page_iterator:
    print("Started Run #" + str(runCount))
    targetBucket = s3.list_objects_v2(Bucket='cloudtrail-logs-cloudthat')
    for element in targetBucket['Contents']:
        if(re.search(pattern,element['Key'])):
            targetFiles.append(element['Key'])
    print("Finished Run #" + str(runCount))
print("Done!")
```
